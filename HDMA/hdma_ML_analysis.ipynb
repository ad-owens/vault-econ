# Feature Engineering and Exploratory

```python
def feature_importance(X, df_y):
    
    y = df_y['action_taken']
    model = ExtraTreesClassifier()
    clf = model.fit(X, y)

    feature_importance = clf.feature_importances_.tolist()
    features = X.columns.tolist()
    
    df_features = pd.DataFrame( {'Feature_Importance':feature_importance,'Features':features})

    df_features = df_features[['Features','Feature_Importance']]
    feature_list = df_features.loc[df_features['Feature_Importance'] > 0.00001]['Features'].tolist()
    
    return feature_list
```

```python
def df_preprocess(df):
    
    print "Original DF Shape:"+str(df.shape)
    
    df_y = df[['action_taken','action_taken_name']]
    df1 = df.drop(['action_taken','action_taken_name','sequence_number','census_tract_number'],axis=1)
    
    #convert rate_spread to binary and interpolate remaining
    df1['spread_given'] = df1['rate_spread'].map(lambda x: 1 if x > 0 else 0)
    
    df1 = df1.drop('rate_spread', axis=1).interpolate()
    
    print "2nd Transformation Shape:"+str(df1.shape)
        
    feature_list = feature_importance(df1, df_y)
    df2 = df1[feature_list]
    
    print "Feature Selection Shape:"+str(df2.shape)
    
    print "df_y:"+str(df_y.shape)
        
    X_train, X_test, y_train, y_test = train_test_split(df2, df_y, test_size=0.20, random_state=25)
    
    print "X_train Shape:"+str(X_train.shape)
    print "X_test Shape:"+str(X_test.shape)
    print "y_train Shape:"+str(y_train.shape)
    print "y_test Shape:"+str(y_test.shape)
    
    return X_train, X_test, y_train, y_test 
```

```python
X_train, X_test, y_train, y_test = df_preprocess(pd.read_csv('hdma_total.csv').sample(n=30000))
```
